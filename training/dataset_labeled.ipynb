```python
from tqdm import tqdm
import os
import shutil
import pandas as pd
from pathlib import Path
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
```

```python
INPUT_DIR = Path("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection/train")

OUTPUT_DIR_TRAIN = Path("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection_labeled/train")
OUTPUT_DIR_VAL = Path("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection_labeled/val")
```

```python
OUTPUT_DIR_TRAIN.mkdir(parents=True, exist_ok=True)
OUTPUT_DIR_VAL.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection/trainLabels.csv")
df.head(10)
```

```python
train_df, val_df = train_test_split(df, test_size=0.2, stratify=df['level'], random_state=42)
```

```python
def organize_dataset(df_split, input_dir, output_dir):
    for _, row in tqdm(df_split.iterrows(), total=len(df_split)):
        img_name = row["image"] + ".jpeg"
        label = str(row['level'])

        dst_dir = output_dir / label
        os.makedirs(dst_dir, exist_ok=True)

        src_path = input_dir / img_name
        dst_path = dst_dir / img_name

        shutil.copy(src_path, dst_path)
```

```python
organize_dataset(train_df, INPUT_DIR, OUTPUT_DIR_TRAIN)
```

```python
organize_dataset(val_df, INPUT_DIR, OUTPUT_DIR_VAL)
```
