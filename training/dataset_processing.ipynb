```python
import cv2
import numpy as np
from pathlib import Path
from tqdm import tqdm
```

```python
INPUT_DIR_TRAIN = Path("/media/juan/Respaldo/diabetic-retinopathy-detection/train/train")
OUTPUT_DIR_TRAIN = Path("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection/train")

INPUT_DIR_TEST = Path("/media/juan/Respaldo/diabetic-retinopathy-detection/test/test")
OUTPUT_DIR_TEST = Path("/home/juan/Documentos/Diabetic Retinopathy Detection/dataset/Dataset_Diabetic_Retinopathy_Detection/test")
```

```python
OUTPUT_DIR_TRAIN.mkdir(parents=True, exist_ok=True)
OUTPUT_DIR_TEST.mkdir(parents=True, exist_ok=True)
```

```python
IMG_SIZE =224
```

```python
def crop_retina(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    _, thresh = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
    
    coords = np.column_stack(np.where(thresh > 0))
    
    if coords.shape[0] == 0:
        return img 
    
    y_min, x_min = coords.min(axis=0)
    y_max, x_max = coords.max(axis=0)
    
    cropped = img[y_min:y_max, x_min:x_max]
    
    return cropped
```

```python
image_paths = list(INPUT_DIR_TRAIN.glob("*.jpeg"))

for img_path in tqdm(image_paths, desc="Processing images"):
    try:

        img = cv2.imread(str(img_path))
        
        if img is None:
            continue

        img = crop_retina(img)

        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

        save_path = OUTPUT_DIR_TRAIN/ img_path.name
        cv2.imwrite(str(save_path), img, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

    except Exception as e:
        print(f"Error with {img_path}: {e}")

print("✅ Dataset processed and ready")
```

```python
image_paths = list(INPUT_DIR_TEST.glob("*.jpeg"))

for img_path in tqdm(image_paths, desc="Processing images"):
    try:

        img = cv2.imread(str(img_path))
        
        if img is None:
            continue

        img = crop_retina(img)

        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

        save_path = OUTPUT_DIR_TEST / img_path.name
        cv2.imwrite(str(save_path), img, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

    except Exception as e:
        print(f"Error with {img_path}: {e}")

print("✅ Dataset processed and ready")
```
